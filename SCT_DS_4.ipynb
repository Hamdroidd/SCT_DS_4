```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
df = pd.read_csv("RTA Dataset.csv")

# Convert Time to datetime
df['Time'] = pd.to_datetime(df['Time'], format='%H:%M:%S', errors='coerce')
df['Hour'] = df['Time'].dt.hour

# 1. Accidents by Hour of Day
plt.figure(figsize=(10, 5))
sns.countplot(x='Hour', data=df, palette='coolwarm', hue='Hour', legend=False)
plt.title('Accidents by Hour of Day')
plt.xlabel('Hour')
plt.ylabel('Number of Accidents')
plt.show()

# 2. Accidents by Day of Week
plt.figure(figsize=(10, 5))
order = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
sns.countplot(x='Day_of_week', data=df, order=order, palette='viridis', hue='Day_of_week', legend=False)
plt.title('Accidents by Day of Week')
plt.xlabel('Day')
plt.ylabel('Number of Accidents')
plt.xticks(rotation=45)
plt.show()

# 3. Accident Severity Distribution
plt.figure(figsize=(8, 4))
sns.countplot(x='Accident_severity', data=df, palette='Set2', hue='Accident_severity', legend=False)
plt.title('Distribution of Accident Severity')
plt.xlabel('Severity')
plt.ylabel('Count')
plt.show()

# 4. Causes of Accidents (Top 10)
plt.figure(figsize=(12, 6))
df['Cause_of_accident'].value_counts().nlargest(10).plot(kind='barh', color='tomato')
plt.title('Top 10 Causes of Accidents')
plt.xlabel('Number of Accidents')
plt.gca().invert_yaxis()
plt.show()

# 5. Accident Severity vs Type of Vehicle
plt.figure(figsize=(12, 6))
sns.countplot(data=df, y='Type_of_vehicle', hue='Accident_severity', palette='magma')
plt.title('Accident Severity by Vehicle Type')
plt.xlabel('Count')
plt.ylabel('Vehicle Type')
plt.legend(title='Severity')
plt.tight_layout()
plt.show()
```
